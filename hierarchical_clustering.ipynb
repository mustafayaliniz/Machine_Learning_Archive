```python
# Veriyi oku
data = pd.read_csv('musteriler.csv')
data
```

```python
# Veriyi bol
X = data.iloc[:,3:].values
```

[Documentation AgglomerativeClustering](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html)

```python
# Veriyi egit ve tahmin et
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
pred = model.fit_predict(X)
```

```python
# Tahmin edilmis siniflarin grafigini goster
from matplotlib import pyplot as plt
plt.scatter(X[pred==0,0],X[pred==0,1],s=100,label='cluster 1')
plt.scatter(X[pred==1,0],X[pred==1,1],s=100,label='cluster 2')
plt.scatter(X[pred==2,0],X[pred==2,1],s=100,label='cluster 3')
plt.title('Agglomerative Clustering',fontweight='bold')
plt.legend()
plt.show()
```

[Documentation scipy.cluster.hierarchy](https://docs.scipy.org/doc/scipy/reference/cluster.hierarchy.html)

```python
# Dendrogram grafigini goster
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(X,metric='euclidean',method='ward')
dendrogram(Z)
plt.show()
```
