```python
#veri csv dosyasi pandas ile okundu
df=pd.read_csv('eksikveriler.csv')
df
```

```python
#NaN degerleri olan sutunlar alindi ortalamasi hesaplandi 
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df[['yas','boy','kilo']].values)
yas=imputer.transform(df[['yas','boy','kilo']].values)
yas # 28.45 olarak Nan degerlerini doldurdu
```

```python
#encoder kutuphaneleri yuklendi
from sklearn.preprocessing import LabelEncoder, OneHotEncoder 
```

```python
df=pd.read_csv('veriler.csv')
ulke=df[['ulke']].values
ulke
```

```python
#labelencoder ile sayisal veriye donusturuyorum
le=LabelEncoder()
data=le.fit_transform(ulke[:,0])
data
```

```python
#tek boyutlu (1,22) matrisi (22,1) ceviriyorum
data=data.reshape(-1,1)
data
```

```python
#onehotencoder kategorik veriyi sayisal veriye donusturuyorum
ohe=OneHotEncoder(categories='auto')
data=ohe.fit_transform(data).toarray()
ulke=data
ulke
```

```python
#veriden sutun alma 
cinsiyet=df[['cinsiyet']].values
cinsiyet
```

```python
#ordinal veriyi label encode ediyoruz
cinsiyet=le.fit_transform(cinsiyet[:,0])
cinsiyet=cinsiyet.reshape(-1,1)
cinsiyet
```

```python
#Dataframe olusturma veriye baslik ekleme 
sonuc=pd.DataFrame(data=yas,columns=['yas','boy','kilo'])
sonuc1=pd.DataFrame(data=cinsiyet,columns=['cinsiyet'])
sonuc2=pd.DataFrame(data=ulke,columns=['fr','tr','us']) 
sonuc2
```

```python
#Dataframeleri birlestirme 
data=pd.concat([sonuc,sonuc2],axis=1)#axis satir sutun belli ediyor
data
```

```python
#veriyi train ve test icin kutuphaneler
from sklearn.model_selection import train_test_split
```

```python
#x bagimsiz degisken y bagimli desgisken yuzde 33  olarak ayirdik butun veriyi
x_train,x_test,y_train,y_test=train_test_split(data,sonuc1,test_size=0.33,random_state=42) 
```

```python
#data olceklendirme kutuphanesi 
from sklearn.preprocessing import StandardScaler
```

```python
#datayi olceklendiriyoruz 
sc=StandardScaler()
X_train=sc.fit_transform(x_train)
X_test=sc.transform(x_test)
X_train
```
