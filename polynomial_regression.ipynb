```python
import pandas as pd
```

```python
# Veriyi oku 
data=pd.read_csv('maaslar.csv')
data
```

```python
# Veriyi Bagimli Bagimsiz degisken olarak ayiriyoruz
X=data[['Egitim Seviyesi']]
Y=data[['maas']]
```

```python
# Simple Linear Regression
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
lr = LinearRegression()
lr.fit(X,Y)
plt.scatter(X,Y,color='red')
plt.plot(X,lr.predict(X))
plt.show()
```

```python
# Polynomial Regression Degree: '2'
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2) # Degree '2'
X_poly=poly.fit_transform(X)
lr.fit(X_poly,Y)
plt.scatter(X,Y,color='red')
plt.plot(X,lr.predict(X_poly))
plt.show()
```

```python
# Polynomial Regression Degree: '4'
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
lr2 = LinearRegression()
poly1 = PolynomialFeatures(degree=4) # Degree '4'
X_poly = poly1.fit_transform(X)
lr2.fit(X_poly,Y)
plt.scatter(X,Y,color='red')
plt.plot(X,lr2.predict(X_poly))
plt.show()
```
