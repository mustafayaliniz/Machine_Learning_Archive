```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Veriyi Okuma
data = pd.read_csv('maaslar.csv')
```

```python
# Veriyi Bolme 
X = data.iloc[:, 1:2].values
Y = data.iloc[:, 2:].values
```

```python
# Model egitme
from sklearn.ensemble import RandomForestRegressor
rfb = RandomForestRegressor(n_estimators=10, random_state=0)
rfb.fit(X, Y.ravel())
```

```python
# Grafik olusturma 
plt.scatter(X, Y , color='red', label='Data')
plt.plot(X, rfb.predict(X), label='Random Forest Prediction')
plt.title('Random Forest Regression')
plt.xlabel('Bagimsiz Degisken')
plt.ylabel('Bagimli Degisken')
plt.legend()
plt.show()
```

```python
# 6.6 degeri icin tahmin
rfb.predict([[6.6]])
```
