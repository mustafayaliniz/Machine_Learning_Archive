```python
# Veriyi oku
data = pd.read_csv('veriler.csv')
data
```

```python
# Veriyi Bagimli bagimsiz degisken olarak bol
X = data.iloc[:, 1:4].values
Y = data.iloc[:, 4].values
```

```python
# Veriyi linear model olarak egit
from sklearn.svm import SVC
svc1 = SVC(kernel='linear')
svc1.fit(X, Y)
```

```python
# Veriyi poly model olarak egit
svc2 = SVC(kernel='poly',degree=2) # Degree: 2 
svc2.fit(X, Y)
```

```python
# Veriyi Radial Basis Function olarak egit
svc3 = SVC(kernel='rbf',gamma=0.1) # egriyi artiriyorum
svc3.fit(X, Y)
```

```python
# Verileri tahminde bulun ve DataFrame olarak kaydet
pred1 = pd.DataFrame(svc1.predict(X), columns=['linear pred'])
pred2 = pd.DataFrame(svc2.predict(X), columns=['poly pred'])
pred3 = pd.DataFrame(svc3.predict(X), columns=['rbf pred'])
```

```python
# DataFrameleri birlestir
df = pd.concat([data,pred1, pred2, pred3], axis=1)
df
```
