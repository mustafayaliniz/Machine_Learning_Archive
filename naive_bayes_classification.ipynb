```python
# Verileri oku 
data = pd.read_csv('veriler.csv')
data
```

```python
# Verileri Bagimli Bagimsiz degisken olarak ayir
X = data.iloc[:,1:4].values
Y = data.iloc[:,4].values
```

```python
# Verileri Gaussian yontemiyle egit
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X,Y)
```

```python
# Verileri Bernoulli yontemiyle egit
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(X,Y)
```

```python
# Verileri multinomial yontemiyle egit
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X,Y)
```

```python
# Veriler uzerinde tahminde bulun ve dataframe olarak kaydet
df1 = pd.DataFrame(gnb.predict(X),columns=['Gaussian Naive Bayes'])
df2 = pd.DataFrame(bnb.predict(X),columns=['Bernoulli Naive Bayes'])
df3 = pd.DataFrame(mnb.predict(X),columns=['Multinomial Naive Bayes'])
frame = pd.concat([data['cinsiyet'],df1,df2,df3],axis=1)
frame
```
