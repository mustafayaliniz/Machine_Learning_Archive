```python
# Veriyi oku 
data = pd.read_csv('Churn_Modelling.csv')
data
```

```python
# Veriyi Bagimli ve bagimsiz degisken olarak bol 
X = data.iloc[:,3:13].values
Y = data.iloc[:,13].values
```

```python
# Veriyi encode et ve sutunlari ayarla 
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
le = LabelEncoder()
X[:,1] = le.fit_transform(X[:,1])
X[:,2] = le.fit_transform(X[:,2])
ohe = ColumnTransformer([("ohe", OneHotEncoder(dtype=float),[1])],remainder="passthrough")
X = ohe.fit_transform(X)
X = X[:,1:]
X
```

```python
# Veriyi test ve train olmak uzere ikiye ayir
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=0)
```

```python
# Veriyi scale et 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Veriyi egit
from keras import models
from keras import layers
model = models.Sequential()
model.add(layers.Dense(6, activation='relu', input_shape=(11,))) # Girdi 11
model.add(layers.Dense(6, activation='relu')) # Ara Katman 
model.add(layers.Dense(1, activation='sigmoid')) # Cikis
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 
history = model.fit(X_train, Y_train, epochs=100, validation_split=0.2, verbose=2) # Veriyi egit 
y_pred = model.predict(X_test) # Tahminde bulun 
y_pred = (y_pred > 0.5) # Tahmin 0.5 usutundeyse True olark ayarla
```
