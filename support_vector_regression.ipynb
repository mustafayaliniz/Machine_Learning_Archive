```python
import pandas as pd
from matplotlib import pyplot as plt
```

@author: MUSTAFA YALINIZ

```python
data=pd.read_csv('maaslar.csv')
data
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X=data.iloc[:,1:2].values
Y=data.iloc[:,2:].values
```

```python
x_olceki = sc.fit_transform(X)
y_olceki = sc.fit_transform(Y)
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='rbf') # Radial Basis Function
svr.fit(x_olceki,y_olceki.ravel())
plt.scatter(x_olceki,y_olceki, color='red')
plt.plot(x_olceki,svr.predict(x_olceki))
plt.title('SVR Radial Basis Function')
plt.xlabel('Bagimsiz Degisken')
plt.ylabel('Bagimli Degisken')
plt.show()
```

```python
svr1=SVR(kernel='linear') # Linear Regression
svr1.fit(x_olceki,y_olceki.ravel())
plt.scatter(x_olceki,y_olceki, color='red')
plt.plot(x_olceki,svr1.predict(x_olceki))
plt.title('SVR Linear Regression')
plt.xlabel('Bagimsiz Degisken')
plt.ylabel('Bagimli Degisken')
plt.show()
```

```python
svr2=SVR(kernel='poly')
svr2.fit(x_olceki,y_olceki.ravel())
plt.scatter(x_olceki,y_olceki, color='red')
plt.plot(x_olceki,svr2.predict(x_olceki))
plt.title('SVR Polynomial Regression')
plt.xlabel('Bagimsiz Degisken')
plt.ylabel('Bagimli Degisken')
plt.show()
```
