```python
# Data okunur
data=pd.read_csv('veriler.csv')
data
```

```python
# Encode kutuphansei yuklenir
from sklearn.preprocessing import LabelEncoder 
le=LabelEncoder()
```

```python
# Encode edilir
data['cinsiyet']=le.fit_transform(data['cinsiyet'])
```

```python
cinsiyet=data[['cinsiyet']]
```

```python
cinsiyet
```

```python
# Sutunun adi degistirilir
erkek=cinsiyet.rename(columns={'cinsiyet':'erkek'})
```

```python
ulke=data[['ulke']]
```

```python
ulke
```

```python
# 3 Ayri sutun olusturulur
ulke=pd.get_dummies(ulke,dtype=int)
```

```python
ulke
```

```python
# Ulke sutunu atilir
data=data.drop(['ulke'],axis=1)
```

```python
# Iki veri birlestirilir
data=pd.concat([ulke,data],axis=1)
data
```

```python
# Bagimli ve bagimsiz degisken bolunur
X=data.drop(['boy'],axis=1)
Y=data[['boy']]
```

```python
# Test Train kutuphanesi yuklenir
from sklearn.model_selection import train_test_split
```

```python
# Veri Test Train olmak uzere ikiye ayrilir
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=0,test_size=0.33)
```

```python
# Egitim kutuphanesi yuklenir
from sklearn.linear_model import LinearRegression
```

```python
# Model egitilir ve tahmin yapilir
lreg=LinearRegression()
lreg.fit(x_train,y_train)
prediction=lreg.predict(x_test)
```

```python
prediction
```

```python
y_test
```

```python
# Hata orani ogrenilir
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,prediction)
mse
```
